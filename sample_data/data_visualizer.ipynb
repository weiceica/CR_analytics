```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv('card_stats.csv')
```

```python
# Bar Chart of Pick Rates
plt.figure(figsize=(15, 8))  # Adjust the figure size as needed
pick_rate_plot = data.sort_values('pick_rate', ascending=False).plot.bar(x='Cards', y='pick_rate', color='blue')
pick_rate_plot.set_title('Pick Rate of Each Card')
pick_rate_plot.set_xlabel('Card')
pick_rate_plot.set_ylabel('Pick Rate')
pick_rate_plot.tick_params(axis='x', rotation=90)  # Rotate labels
plt.tight_layout()
plt.show()
```

```python
# Bar Chart of Win Rates
plt.figure(figsize=(10, 6))
data.sort_values('win_rate', ascending=False).plot.bar(x='Cards', y='win_rate', color='green')
plt.title('Win Rate of Each Card')
plt.xlabel('Card')
plt.ylabel('Win Rate')
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot of Wins vs. Played
plt.figure(figsize=(10, 6))
plt.scatter(data['played'], data['wins'], alpha=0.5)
plt.title('Wins vs. Played for Each Card')
plt.xlabel('Number of Times Played')
plt.ylabel('Number of Wins')
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('Golem_data.csv')

data['Date'] = pd.to_datetime(data['Date'], format='%Y%m%d')

# Sort the DataFrame based on the date
df = data.sort_values(by='Date')

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Win Rate'], marker='o')  # 'o' adds dots at each data point
plt.title('Win Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Win Rate')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.grid(True)
plt.show()
```

```python
data = pd.read_csv('Electro Giant_data.csv')

data['Date'] = pd.to_datetime(data['Date'], format='%Y%m%d')

# Sort the DataFrame based on the date
df = data.sort_values(by='Date')

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Pick Rate'], marker='o')  # 'o' adds dots at each data point
plt.title('Pick Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Pick Rate')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['Win Rate'], df['Pick Rate'], alpha=0.5)  # 'o' adds dots at each data point
plt.title('Win Rate vs Pick Rate')
plt.xlabel('Win Rate')
plt.ylabel('Pick Rate')
plt.grid(True)
plt.show()
```
