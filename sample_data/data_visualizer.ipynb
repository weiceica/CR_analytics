```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv('card_stats.csv')  # Replace with your CSV file path
```

```python
# Bar Chart of Pick Rates
plt.figure(figsize=(15, 8))  # Adjust the figure size as needed
pick_rate_plot = data.sort_values('pick_rate', ascending=False).plot.bar(x='Cards', y='pick_rate', color='blue')
pick_rate_plot.set_title('Pick Rate of Each Card')
pick_rate_plot.set_xlabel('Card')
pick_rate_plot.set_ylabel('Pick Rate')
pick_rate_plot.tick_params(axis='x', rotation=90)  # Rotate labels
plt.tight_layout()
plt.show()
```

```python
import json

# Step 1: Read the original JSON file
with open('detailed_card_info.json', 'r', encoding='utf-8') as file:
    original_data = json.load(file)

# Step 2: Transform the data
new_data = {}
for item in original_data:
    # Assuming each item is a dictionary and has an 'id' key
    item_id = item['id']
    new_data[item_id] = item['name']

# Step 3: Write the new JSON file
with open('card_info.json', 'w') as file:
    json.dump(new_data, file, indent=4)
```
