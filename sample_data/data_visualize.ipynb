```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv('card_stats.csv')
```

We First plot a bar chart of how often the cards were picked for some initial visualization

```python
'''Sort data based on Pick rate + Split Data into 3 tiers of picked cards for visualization'''
data_sorted = data.sort_values('pick_rate', ascending=False)
chunks = np.array_split(data_sorted, 3)
```

```python
'''plot the first chunk'''
plt.figure(figsize=(20, 7))
plt.subplot(1, 3, 1)
plt.bar(chunks[0]['Cards'], chunks[0]['pick_rate'], color='blue')
plt.title('Pick Rate of Each Card (Part 1)')
plt.xlabel('Card')
plt.ylabel('Pick Rate')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
'''plot the second chunk'''
plt.figure(figsize=(20, 7))
plt.subplot(1, 3, 1)
plt.bar(chunks[1]['Cards'], chunks[1]['pick_rate'], color='blue')
plt.title('Pick Rate of Each Card (Part 2)')
plt.xlabel('Card')
plt.ylabel('Pick Rate')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
'''plot the third chunk'''
plt.figure(figsize=(20, 7))
plt.subplot(1, 3, 1)
plt.bar(chunks[0]['Cards'], chunks[0]['pick_rate'], color='blue')
plt.title('Pick Rate of Each Card (Part 3)')
plt.xlabel('Card')
plt.ylabel('Pick Rate')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('Golem_data.csv')

data['Date'] = pd.to_datetime(data['Date'], format='%Y%m%d')

# Sort the DataFrame based on the date
df = data.sort_values(by='Date')

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Win Rate'], marker='o')  # 'o' adds dots at each data point
plt.title('Win Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Win Rate')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.grid(True)
plt.show()
```

```python
data = pd.read_csv('Electro Giant_data.csv')

data['Date'] = pd.to_datetime(data['Date'], format='%Y%m%d')

# Sort the DataFrame based on the date
df = data.sort_values(by='Date')

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Pick Rate'], marker='o')  # 'o' adds dots at each data point
plt.title('Pick Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Pick Rate')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['Win Rate'], df['Pick Rate'], alpha=0.5)  # 'o' adds dots at each data point
plt.title('Win Rate vs Pick Rate')
plt.xlabel('Win Rate')
plt.ylabel('Pick Rate')
plt.grid(True)
plt.show()
```
